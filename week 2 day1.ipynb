# ploting a piechart 

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
plt.pie(x)
plt.show()
```

# To label this pie chart label is used 

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple","Mango","kiwi","Oragne"]
plt.pie(x,labels=my_label)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple","Mango","kiwi","Oragne"]
plt.pie(x,labels=my_label,startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple","Mango","kiwi","Oragne"]
plt.pie(x,labels=my_label,startangle=180)
plt.show()
```

# partition : separating all of the other to show the differention my_explode method is used   

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple","Mango","kiwi","Oragne"]
my_explode=[0.2,0,0,0]
plt.pie(x,labels=my_label,explode=my_explode)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([35,25,25,15])
my_label=["Apple","Mango","kiwi","Oragne"]
my_explode=[0.2,0.2,0.2,0.2]
plt.pie(x,labels=my_label,explode=my_explode)
plt.show()
```
